```python
import math
import matplotlib.pyplot as plt
```

```python
N=500
x=[]
y_sin=[]
y_sin_1appx=[]
y_sin_3appx=[]
for i in range(N):
    theta=(2*math.pi/N)*i-(math.pi/2)
    x.append(theta)
    y_sin.append(math.sin(theta))
    y_sin_1appx.append(theta)
    y_sin_3appx.append(theta-(1/6)*theta**3)
    
```

```python
plt.plot(x,y_sin,
         label="$\\sin \\theta$")
plt.plot([min(x),max(x)],[0,0],
         label="0th order, $sin(\\theta) \\simeq 0$")
plt.plot(x,y_sin_1appx,         
         label="1st order, $sin(\\theta) \\simeq \\theta$")
plt.plot(x,y_sin_3appx,         
         label="3rd order, $sin(\\theta) \\simeq \\theta-\\frac{1}{6} \\theta^3$")
plt.grid()
plt.ylabel("sine or approximation to sine")
plt.xlabel("theta (radians)")
plt.legend()
```

```python
# compute error in 3rd order approximation
abs_error=[]
for i in range(len(x)):
    abs_error.append( abs((y_sin_3appx[i]-y_sin[i])))

plt.plot(x,abs_error,
         label="$ | ( \\theta -\\frac{1}{6} \\theta^3) - sin \\theta  | $")
plt.grid()
plt.ylabel("$ | ( \\theta -\\frac{1}{6} \\theta^3) - sin \\theta | $")
plt.xlabel("theta (radians)")
plt.legend()
```
