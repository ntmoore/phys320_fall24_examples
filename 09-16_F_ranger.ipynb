```python
G=6.67e-11
R_earth_moon=3.84e8 # meters, https://en.wikipedia.org/wiki/Lunar_distance
m_earth=5.97219e24 # kg
m_moon=7.34767309e22 # kg
r_earth=6.371e6 # meters
r_moon= 1.74e6 # meters
```

```python
# Force of gravity on 200km above Earth's surface
G*m_earth*1/(r_earth+200e3)**2
```

```python
# force of gravity at 50km up from earth's surface
G*m_earth*1/(50e3+r_earth)**2
```

```python
r_vals=[]
F_vals=[]
r=r_earth+50e3
dr = 1000e3 # 10km steps of distance
for i in range(1,10):
    r_vals.append(r)
    F_vals.append(G*m_earth*1/r**2)
    r = r + dr
```

```python
#import numpy as np 
import matplotlib.pyplot as plt
```

```python
plt.plot(r_vals,F_vals)
plt.ylabel("Force on 1kg mass (N)")
plt.xlabel("distance from Earth's center (m), 6.4e6 m radius")
plt.title("Earth's Gravitational Pull, dR=%.1e m steps"%dr)
plt.grid()
```

```python
r_vals=[]
F_e_vals=[]
F_m_vals=[]
F_net_vals=[]
r=r_earth
N=1000
dr = (R_earth_moon-r_earth-r_moon)/N
for i in range(N):
    r=r_earth + i*dr
    r_vals.append(r)
    F_e = -G*m_earth*1/r**2
    F_m = G*m_moon*1/(R_earth_moon-r)**2
    F_net = F_e+F_m
    F_e_vals.append(F_e)
    F_m_vals.append(F_m)
    F_net_vals.append(F_net)
                
```

```python
plt.plot(r_vals,F_e_vals)
plt.plot(r_vals,F_m_vals)
plt.plot(r_vals,F_net_vals)
plt.ylabel("Force on 1kg mass (N)")
plt.xlabel("distance from Earth's center (m), 6.4e6 m radius")
plt.title("Earth's Gravitational Pull, dR=%.1e m steps"%dr)
plt.grid()
```
