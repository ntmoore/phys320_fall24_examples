```python
import math
import matplotlib.pyplot as plt
```

```python
N=500
x=[]
y_sin=[]
y_sin_appx=[]
for i in range(N):
    theta=(2*math.pi/N)*i-(math.pi/2)
    f_sine=math.sin(theta)
    f_appx_sin=theta-(1/6)*theta**3
    x.append(theta)
    y_sin.append(f_sine)
    y_sin_appx.append(f_appx_sin)
    
```

```python
plt.plot(x,y_sin)
plt.plot(x,y_sin_appx)
plt.grid()
plt.ylabel("sine or approximation to sine")
plt.xlabel("theta (radians)")
```
