```python
#Rocket Problem
G=6.67e-11
r_earth_moon=3.84e8 #m
m_earth=5.972e24 #kg
m_moon=7.348e22 #kg
r_earth=6380e3 + 5e4 #m
r_moon=174e4 #m
m_rocket=365.5 #kg
v0=14000 #m/s
```

```python
#Force on Rocket
F_earth=-G*((m_earth*m_rocket)/(r_earth**2))
F_moon=G*((m_moon*m_rocket)/(r_earth_moon-r_moon**2))
F_net=F_earth+F_moon
Ue=(G*((m_earth*m_rocket)/(r2)))
Um=(G*((m_moon*m_rocket)/(r_earth_moon-r2)))
Unet=Ue+Um
vmin=((2*Unet)/m_rocket)**0.5
print("Force on rocket by Earth:",F_earth)
print("Force on rocket by Moon:",F_moon)
print("Net force on rocket:",F_net)
print("Total Potential Energy",Unet)
print("Minimum escape velocity",vmin)
```

```python
t=0
t_limit = 2000000
v=v0
v_s=[]
r_s=[]
Fn_s=[]
t_s=[]
Un_s=[]
Ke_s=[]
En_s=[]
dt=1.1
a1=-F_net/m_rocket
r1=r_earth
r2=r1 + v0*dt + 0.5*a1*dt*dt
while (r2<(r_earth_moon-r_moon) and t<t_limit and r2>r_earth):
    F_e = (-G*(m_earth*m_rocket)/(r2**2))
    F_m =((G*m_moon*m_rocket)/(r_earth_moon-r2)**2)
    Fnet = F_e + F_m
    Ue=(G*((m_earth*m_rocket)/(r2)))
    Um=(G*((m_moon*m_rocket)/(r_earth_moon-r2)))
    Unet=Ue+Um
    Ke=0.5*m_rocket*(v2)**2
    Enet=Ke+Unet
    a2 = -Fnet/m_rocket
    r3 = 2*r2 - r1 + a2*dt*dt
    v2 = (r3-r1)/(2*dt)
    v_s.append(v2)
    r_s.append(r2)
    Fn_s.append(Fnet)
    t_s.append(t)
    Un_s.append(Unet)
    Ke_s.append(Ke)
    En_s.append(Enet)
    t = t + dt
    r1 = r2
    r2 = r3
```

```python
import matplotlib.pyplot as plt
plt.plot(r_s,Fn_s,label="Net Force")
plt.ylabel("Force on Rocket (N)")
plt.xlabel("Distance from Earth (m)")
plt.legend()
plt.grid()
```
