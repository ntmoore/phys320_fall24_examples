```python
import csv
```

```python
x=[]
y=[]
with open("blade_data.csv") as csvfile:
    datafile=csv.reader(csvfile)
    for row in datafile:
        #print(row)
        x.append(row[0])
        y.append(row[1])
print(x)
print(y)
```

```python
len(x)
```

```python
x[0]
```

```python
# throw away the first line
x=x[1:]
y=y[1:]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(x,y)
```

```python
xf=[]
yf=[]
for a in x:
    xf.append(-1*float(a))

for a in y:
    yf.append(float(a))
```

```python
plt.plot(xf,yf,"o", fillstyle='full')
plt.xlabel("shaft direction (inch)")
plt.ylabel("blade direction (inch)")

plt.xlim(-0.5,19)
plt.ylim(-4,4)

plt.xticks(ticks=np.arange(0,19,2),minor=False)
plt.xticks(ticks=np.arange(0,19,0.5),minor=True)
plt.yticks(ticks=np.arange(-4,4,1),minor=False)
plt.yticks(ticks=np.arange(-4,4,0.25),minor=True)

plt.grid(visible=True, which='major', axis='x',linestyle="-",color="black")
plt.grid(visible=True, which='minor', axis='x',linestyle="--",color="green")
plt.grid(visible=True, which='major', axis='y',linestyle="-",color="black")
plt.grid(visible=True, which='minor', axis='y',linestyle="--",color="green")
```
